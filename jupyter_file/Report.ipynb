```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt # Matlab-style plotting
import seaborn as sns
color = sns.color_palette()
sns.set_style('darkgrid')
from scipy import stats
from scipy.stats import norm, skew #for some statistics
```

```python
raw_df = pd.read_csv('data.csv', header = None)
```

```python
def re_cate(label):
    if label >2:
        return 1
    else:
        return 0
```

```python
raw_df.rename(columns = {0:'Ae', 1:'Ap', 2:'Rs', 3:'Fz', 4:'label'}, inplace = True)
raw_df['label'] = raw_df.apply(lambda x : re_cate(x.label), axis = 1)
```

```python
raw_df
```

```python
fig, ax = plt.subplots()
ax.scatter(x = raw_df['Rs'], y = raw_df['label'])
plt.ylabel('Quality', fontsize=13)
plt.xlabel('Rs', fontsize=13)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x = raw_df['Ap'], y = raw_df['label'])
plt.ylabel('Quality', fontsize=13)
plt.xlabel('Ap', fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(10,10))
g = sns.heatmap(raw_df.corr(),annot=True,cmap="RdYlGn")
```
